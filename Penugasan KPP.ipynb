## Analisis Data
Untuk data yang ingin dianalisis, saya mengambil kolom "start_station_name" dan "end_station_name" untuk menghitung jumlah pemnijaman di tiap - tiap stasiun

```python
df_selected = df.iloc[:,[4,6]] # untuk mengambil data frame dengan kolom "start_station_name" dan end_station_name
print(df_selected)
```

```python
start_counts = df_selected['start_station_name'].value_counts()  # Menghitung jumlah pada kolom start_station_name
end_counts = df_selected['end_station_name'].value_counts()  # Menghitung jumlah pada kolom end_station_name

print(start_counts)
print ()
print(end_counts)
```

```python
start=df['start_station_name'].value_counts().reset_index(name='count_start').rename(columns={'index': 'start_station_name'}) # Membuat kolom baru bernama count_start
end=df['end_station_name'].value_counts().reset_index(name='count_end').rename(columns={'index': 'end_station_name'}) # Membuat kolom baru bernama count_start

print(start)
print(end)
```

```python
# Menyimpan DataFrame start ke CSV
start.to_csv('start_stations.csv', index=False)

# Menyimpan DataFrame end ke CSV
end.to_csv('end_stations.csv', index=False)  # Menyimpan DataFrame ke file CSV
```
